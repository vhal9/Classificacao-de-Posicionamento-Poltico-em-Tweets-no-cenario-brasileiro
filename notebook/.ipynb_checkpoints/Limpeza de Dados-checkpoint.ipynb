# Limpeza dos dados
- remover tweets que n√£o est√£o em portugu√™s

```python
import pandas as pd
```

```python
database = pd.read_csv('../data/corpus.csv', usecols=['id','lang', 'text', 'url1', 'url2'])
```

## Remover tweets que n√£o est√£o em portugu√™s

```python
dataframe = database.loc[(database["lang"]=="pt"),['id','lang', 'text', 'url1', 'url2']]
```

```python
dataframe
```

## Removendo duplicatas

```python
dataFrame = dataframe.drop_duplicates(dataframe.columns[~dataframe.columns.isin(['text'])],
                        keep='first')
```

```python
dataFrame
```

```python
dataFrame = dataFrame.drop('lang',axis=1)
```

```python
dataFrame = dataFrame.drop('url1', axis=1)
```

```python
dataFrame = dataFrame.drop('url2', axis=1)
```

```python
dataFrame.to_csv('../data/dados.csv')
```
