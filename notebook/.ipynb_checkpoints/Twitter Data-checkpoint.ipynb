```python
import os
import tweepy as tw
```

```python
consumer_key = ''
consumer_secret = ''
```

```python
access_token = ''
access_token_secret = ''
```

```python
auth = tw.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_token_secret)
api = tw.API(auth, wait_on_rate_limit=True)
```

# Lista de usuários para pesquisa

```python
listOfInfluencersEsquerda = ['Haddad_Fernando', 'GuilhermeBoulos','cirogomes', 'LulaOficial', 'gduvivier', 'ManuelaDavila', 'MarceloFreixo', 'psol50', 'samiabomfim', 'lucianagenro', 'randolfeap']
listOfInfluencersDireita = ['jairbolsonaro', 'kimpkat', 'arthurmoledoval', 'moura_101','joaoamoedonovo', 'FernandoHoliday', 'DaniloGentili', 'PSL_Nacional', 'majorolimpio', 'Francischini_', 'OdeCarvalho']
listOfInfluencers = []
listOfInfluencers.extend(listOfInfluencersDireita)
listOfInfluencers.extend(listOfInfluencersEsquerda)
```

```python
len(listOfInfluencers)
```

# Extrair follows dos usuários levantados

```python
foll = []
for user in listOfInfluencers:
    print(user)
    followers = api.followers(id=user, count=40)
    print(len(followers))
    for follower in followers:
#         print(follower.screen_name)
        if follower.protected == False: foll.append(follower.screen_name)
```

```python
listOfInfluencers.extend(foll)
```

```python
len(listOfInfluencers)
```

```python
arquivo = open('../data/users.txt', 'w')
for user in listOfInfluencers:
    escrita = user + '\n'
    arquivo.write(escrita)
```

```python
arquivo.close()
```

# Extrair Tweets dos usuários

```python
tweets = []
import time
invalidUsers = []
for user in listOfInfluencers:
    usuarios = user
    try:
        data = api.get_user(user)
        if data.protected == False:
            tweets.extend(api.user_timeline(id=user, tweet_mode="extended"))
    except:
        invalidUsers.append(user)
```

```python
import pandas as pd
```

```python
dataframe = {'id':[], 'text':[], 'url1':[], 'url2':[], 'lang':[]}
```

```python
for tweet in tweets:
    dataframe['id'].append(tweet.id)
    dataframe['text'].append(tweet.full_text)
    listUrl1 = []
    listUrl2 = []
    print(tweet.full_text)
    if tweet.entities['urls'] != []:
        listUrl1.append(tweet.entities['urls'][0]['url'])
    if 'media' in tweet.entities and tweet.entities['media'] != 0:
        listUrl2.append(tweet.entities['media'][0]['url'])
    dataframe['url1'].append(listUrl1)
    dataframe['url2'].append(listUrl2)
    dataframe['lang'].append(tweet.lang)
```

```python
print(dataframe['id'][0], dataframe['text'][0], dataframe['url1'][0],dataframe['url2'][0], dataframe['lang'][0])
```
