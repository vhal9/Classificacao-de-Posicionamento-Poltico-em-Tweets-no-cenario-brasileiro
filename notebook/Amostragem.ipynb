# Gerar uma amostragem de tweets para etiquetamento.

```python
import pandas as pd
import numpy as np
```

```python
dataframe = pd.read_csv('../data/dadosNormalizados.csv', usecols=['id', 'normalize'])
```

```python
dataframe
```

```python
df_indexs = list(dataframe.index)
```

## Vamos pegar um numero de indices aleat√≥rios no dataframe
- 20% dos indices = 1387
- extrair o dataframe de treinamento


```python
indexs = np.random.choice(df_indexs,1387,replace=False)
```

```python
df_train = dataframe.iloc[indexs]
```

```python
df_train
```

```python
df_train.to_csv('../data/trainset.csv')
```
